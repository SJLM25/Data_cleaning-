```python
#Lectura del csv y visualizaciÃ³n
import pandas as pd

file_path = '03_Experimento_3-spreadsheet(v1).csv'
df = pd.read_csv(file_path)

df.head(7)  
```

```python
# Separar el dataframe con las variables independientes
df_variables = df.iloc[:5]
df_variables 
```

```python
# Separar el dataframe con los reporteros y corridas finales
df_finales = df.iloc[5:]
new_df_finales = df_finales.drop(columns=df_finales.columns[0])
new_df_finales
```

```python
# Elimina las columnas con valores nulos, ademas de transponer el dataframe
df_var_clean = df_variables.dropna(axis=1, how='all').T

#Se asigna la fila 1 como el nombre de las columnas y elimina esta primera columna
df_var_clean.columns = df_var_clean.iloc[0]
DF_variables = df_var_clean.drop(df_var_clean.index[0])
DF_variables.reset_index(drop=True, inplace=True)

DF_variables
```

```python

# Guarda los valores de la segunda fila en un arreglo 
valores = new_df_finales.iloc[1].values

# Guarda una matriz de n filas por 7 columnas tomando como referencia los valores de la segunda fila que en este caso con los reporteros
valores_matriz = valores.reshape(-1, 7)

# Convertimos la matriz generada en un dataframe, agregando nombre a las variables
df_finales_rs = pd.DataFrame(valores_matriz)
column_names = [
    "step","num_cultures", "bigger-region", "prob_interaction_i",
    "prob_interaction_VF", "probability_copy", "per_information_VF"
]
df_finales_rs.columns = column_names

#Elimina la columna 0 o step
DF_finales = df_finales_rs.drop(columns=df_finales_rs.columns[0])

DF_finales
```

```python
# Combina los dataframes generados
df_combinado = pd.concat([DF_variables, DF_finales], axis=1)  # Combina por columnas
df_combinado.to_csv('03_Experimento_3-DataBase.csv', index= False)
df_combinado
```
