```python
#Lectura del csv y visualización
import pandas as pd

file_path = '03_Experimento_1-spreadsheet(v2).csv'
df = pd.read_csv(file_path)

df.head(7)  
```

```python
file_path_2 = '03_Experimento_1-spreadsheet(v1).csv'
df_2 = pd.read_csv(file_path_2)

df_2.head(7)  
```

```python
df_variables = df_2.iloc[:5]
df_variables
```

```python
df_2_clean = df_variables.dropna(axis=1, how='all').T

#Se asigna la fila 1 como el nombre de las columnas y elimina esta primera columna
df_2_clean.columns = df_2_clean.iloc[0]
dataframe_2 = df_2_clean.drop(df_2_clean.index[0])
dataframe_2.reset_index(drop=True, inplace=True)
dataframe_2
```

```python
# Elimina las columnas con valores nulos, ademas de transponer el dataframe
data_frame = df.dropna(axis=1, how='all')

data_frame
```

```python

valores = data_frame.values
# Guarda una matriz de n filas por 7 columnas tomando como referencia los valores de la segunda fila que en este caso con los reporteros
valores_matriz = valores.reshape(-1, 7)
column_names = [
    "step","num_cultures", "bigger-region", "prob_interaction_i",
    "prob_interaction_VF", "probability_copy", "per_information_VF"
]


df_reshape = pd.DataFrame(valores_matriz)
df_reshape.columns = column_names

df_reshape
```

```python
import numpy as np 
df2_repeated = pd.DataFrame(np.tile(dataframe_2.values, (len(df_reshape) // len(dataframe_2) + 1, 1)), columns=dataframe_2.columns).iloc[:len(df_reshape)]
```

```python
df_combinado = pd.concat([df2_repeated, df_reshape], axis=1)
df_combinado.to_csv('02_Experimento_1-DataBase_2.csv', index= False)
df_combinado
```
